## QUES 1
. Read Total profit of all months and show it using a line plot. Total profit data 
provided for each month. Generated line plot must include the following proper∆ües:
a) X label name = Month Number 
b) Y label name = Total profit

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
plt.plot(df['month_number'], df['total_profit'], marker='o', color='blue', label='Total Profit')

# Adding labels and title
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit for Each Month')
# Adding grid for better readability
plt.grid(True)
# Display the legend
plt.legend()
# Show the plot
plt.show()
```

## QUES 2
Get total profit of all months and show line plot with the following Style proper∆ües.
Total profit data provided for each month (same as 1st ques∆üon). Generated line plot
must include following Style proper∆ües:
a) Line Style do∆©ed and Line-color should be red 
b) Show legend at the lower right loca∆üon.
c) X label name = Month Number 
d) Y label name = Sold units number 
e) Add a circle marker. 
f) Line marker color as read 
g) Line width should be 3 

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
plt.plot(df['month_number'], df['total_profit'], linestyle='dotted', color='red', marker='o', markersize=8, markerfacecolor='red', label='Total Profit', linewidth=3)

# Adding labels and title
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit for Each Month')

# Adding legend at the lower right location
plt.legend(loc='lower right')

# Display the plot
plt.show()
```

## QUES 3
Display the number of units sold per month for each product using mul∆üline plots.
(i.e., Separate Plotline for each product). 

```python
plt.figure(figsize=(12, 8))  # Adjust the figure size if needed

# Iterate over each product column and plot a line for each
for product in df.columns[1:7]:  # Assuming columns 1 to 6 represent the products
    plt.plot(df['month_number'], df[product], label=product)

# Adding labels and title
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.title('Number of Units Sold per Month for Each Product')

# Adding legend
plt.legend()

# Display the plot
plt.show()
```

## QUES 4
. Read toothpaste sales data of each month and show it using a sca∆©er plot. Also, add a
grid in the plot, gridline style should be ‚Äú‚Äì‚Äú

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
plt.scatter(df['month_number'], df['toothpaste'], label='Toothpaste Sales', color='green')

# Adding labels and title
plt.xlabel('Month Number')
plt.ylabel('Toothpaste Sales')
plt.title('Toothpaste Sales per Month')

# Adding grid with gridline style '-'
plt.grid(True, linestyle='--')

# Display the plot
plt.show()
```

## QUES 5
. Read face cream and facewash product sales data and show it using the bar chart. The 
bar chart should display the number of units sold per month for each product. Add a 
separate bar for each product in the same chart.

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
bar_width = 0.4  # Adjust the bar width if needed

# Bar positions for each month
bar_positions_facecream = df['month_number'] - bar_width/2
bar_positions_facewash = df['month_number'] + bar_width/2

# Plotting bars for face cream
plt.bar(bar_positions_facecream, df['facecream'], width=bar_width, label='Face Cream', color='blue')

# Plotting bars for facewash
plt.bar(bar_positions_facewash, df['facewash'], width=bar_width, label='Face Wash', color='orange')

# Adding labels and title
plt.xlabel('Month Number')
plt.ylabel('Number of Units Sold')
plt.title('Face Cream and Face Wash Sales per Month')

# Adding legend
plt.legend()

# Display the plot
plt.show()
```
